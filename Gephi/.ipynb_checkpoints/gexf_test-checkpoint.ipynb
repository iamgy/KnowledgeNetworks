```python
import pandas as pd
from gexf import Gexf
import numpy as np
```

```python
data = pd.read_csv('../knetwork/collaboration_data.csv')
```

```python
data
```

```python
graph_file = Gexf('NPM','a test graph')
```

```python
graph = graph_file.addGraph('directed','dynamic','a test for a particular year')
```

```python
data.iloc[0]
```

```python
data
```

```python
collabs = [col for col in data.columns if '&' in col]
collabs = [s.replace('.1','') for s in collabs]
```

```python
data[collabs]
```

```python
not_collabs =  [col for col in data.columns if '&' not in col]
```

```python
data[not_collabs]
```

```python
num = 0 # this should be the index of the year - we will have to do this one at a time
graph = graph_file.addGraph('directed','dynamic','graph for year '+str(data.iloc[num]['Year']))
collab_data = data.iloc[num][collabs]
pub_data = data.iloc[num][not_collabs]
for i in range(0,len(pub_data)):
    graph.addNode(not_collabs[i],not_collabs[i])
    graph.addNodeAttribute('number',defaultValue=data.iloc[num][not_collabs[i]],type='integer',mode='dynamic',force_id=not_collabs[i])
for j in range(0,len(collab_data)):
    graph.addEdge(collabs[j],collabs[j][0:collabs[j].index('&')-1],collabs[j][collabs[j].index('&')+2:len(collabs[j])+1],weight=collab_data[j])
```

```python
output_file=open("graph_year_2000.gexf","w")
graph_file.write(output_file)
```
